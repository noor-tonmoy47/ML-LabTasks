```python
from numpy import loadtxt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Classroom/2019-20: ML/nn_data.txt', header=None)
dataset.shape
print(dataset.head())
```

```python
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,8]

X.head()
```

**Data Normalize(Scaling)**

```python
cols_to_be_scale = [0,1,2,3,4,5,6,7]

scaler = preprocessing.MinMaxScaler()

X[cols_to_be_scale] = scaler.fit_transform(X[cols_to_be_scale])

X.head()
```

```python
model = Sequential()
model.add(Dense(8, input_shape=(8,), activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=350, batch_size=10)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.3f' % (accuracy*100))
```

```python
predictions = (model.predict(X) > 0.5).astype(int)
for i in range(5):
	print(' %d (expected %d)' % ( predictions[i], y[i]))
```
